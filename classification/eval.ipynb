```python
ret_total = [0.416, 0.404, 0.4394, 0.416] # init with DT results
def ret_eval_res(y_pred, y_true):
    # 计算混淆矩阵
    conf_matrix = confusion_matrix(y_true, y_pred)

    # 计算每个类别的准确率
    class_accuracies = {}
    # for i in range(len(conf_matrix)):
    #     class_accuracies[i] = conf_matrix[i, i] / np.sum(conf_matrix[i, :])
    # print(f"Accuracy of class {class_accuracies}, avg acc {accuracy_score(y_true, y_pred)}")
    precision = precision_score(y_true, y_pred, average='macro')  # None表示计算每个类别的分数
    recall = recall_score(y_true, y_pred, average='macro')
    print(accuracy_score(y_true, y_pred))
    ret_total.extend([accuracy_score(y_true, y_pred)])
    print(f"Precision: {precision}, Recall: {recall}")
```

```python
import Orange
from Orange.evaluation.scoring import compute_CD, graph_ranks
import numpy as np

# 示例数据：每一列代表一个分类器的性能，每一行代表一个数据集的结果 4行8列
results = np.array(ret_total).reshape(-1, 4).transpose()
names = ["DT", 'TS2Vec', "PatchTST", "Sequencer", "OmniScaleCNN", "gMLP", "FormerTime"]
# 计算排名
ranks = np.zeros_like(results)
for i in range(results.shape[0]):
    ranks[i] = len(results[i]) - np.argsort(np.argsort(results[i]))
# 计算平均排名
avg_ranks = np.mean(ranks, axis=0)
print(avg_ranks)
# 计算临界差异 (CD)
cd = compute_CD(avg_ranks, n=len(results), alpha='0.05', test='nemenyi')
graph_ranks(avg_ranks, names=names, cd=cd, width=6, textspace=1.3, reverse=True)
plt.tight_layout()
plt.savefig('cd_ranks.pdf', bbox_inches='tight')
```
